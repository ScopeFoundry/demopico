```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Path to the HDF5 file
h5_file_path = '240807_150221_pico_datalog.h5'

with h5py.File(h5_file_path, 'r') as h5f:
    print(dict(h5f))
    # Navigate to the measurement group
    meas_group = h5f['/measurement/pico_datalog']
    
    # Read the dataset
    pr_data = meas_group['pr_data'][:]
    
    # Extract time and pr values
    time_data = np.array(pr_data[:, 0])
    pr_values = np.array(pr_data[:, 1])
```

```python
# Plot the data
plt.figure()
plt.plot(time_data, pr_values, marker='o', linestyle='-', color='b', label='PR Value')
plt.xlabel('Time (s)')
plt.ylabel('PR Value')
plt.title(' Pico Data Logging Measurement')
plt.legend()
plt.grid(True)
plt.show()
```
